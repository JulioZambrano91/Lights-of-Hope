### Comparacion de graficas entre periodo 2016-2018 y 2019-2021

#### 1. Leemos ambos archivos

```python
import pandas as pd
import matplotlib.pyplot as plt

database = pd.read_excel("../database/data_no_procesada/database_2010_2024.xlsx")

database = database[database["Disaster Subgroup" == "Hydrological"]]

periodo_2022_2024 = database[database["Start Year"] >= 2022]
periodo_2022_2024 = periodo_2022_2024[periodo_2022_2024["Start Year"] <= 2024]

periodo_2019_2021 = database[database["Start Year"] >= 2019]
periodo_2019_2021 = periodo_2019_2021[periodo_2019_2021["Start Year"] <= 2021]

periodo_2016_2018 = database[database["Start Year"] >= 2016]
periodo_2016_2018 = periodo_2016_2018[periodo_2016_2018["Start Year"] <= 2018]
```

```python
periodo_2016_2018.info()
```

#### 2. Comparamos los desastre por region de ambos periodos

```python
region1 = periodo_2016_2018['Region']

region2 = periodo_2019_2021["Region"]

region3 = periodo_2022_2024["Region"] 

plt.figure(figsize=(14,8))

plt.subplot(131)
plt.hist(region1)
plt.ylabel("Cant. de desastres")
plt.xlabel("2016 - 2018")

plt.subplot(132)
plt.hist(region2)
plt.xlabel("2019 - 2021")

plt.subplot(133)
plt.hist(region3)
plt.xlabel("2022 - 2024")


plt.suptitle("Desastre por periodo")
```

La region mas afectada por el tipo de desastre hydrological es Asia y con una amplia diferencia

#### 3. Desastres por periodo de tiempo

```python
region1 = periodo_2016_2018['Start Year']

region2 = periodo_2019_2021["Start Year"]

region3 = periodo_2022_2024["Start Year"] 

r = pd.DataFrame({"r1":region1, "r2":region2, "r3":region3})

plt.hist(r)
plt.title("Desastres por año")
plt.ylabel("Cant. de desastres")
plt.xlabel("Año")
plt.show()
```

el año que mas desastre de este tipo hubo fue 2019

### 4. Muertos | beta

```python
region1 = periodo_2016_2018['Total Deaths']

region2 = periodo_2019_2021['Total Deaths']

region3 = periodo_2022_2024['Total Deaths']

r = pd.DataFrame({"r1":region1, "r2":region2, "r3":region3})

plt.figure(figsize=(14,8))

plt.plot(r, marker='D',linewidth=2, markersize=5)
plt.title('Total de muertos', fontsize=16)
plt.xlabel('Cant. de datos', fontsize=14)
plt.ylabel('Cant. de Muertos', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```
