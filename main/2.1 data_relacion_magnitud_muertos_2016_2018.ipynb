# Impacto de desastres hidrologicos: Magnitud y victimas (2016-2018)

**Importamos la base de datos procesada lista para nuestro analisis**

```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/database_procesada_2010_2024.xlsx')

df.head()
```

Hacemos un limpieza de filas filtrando solamente la columna de muertos, debido que, ya la base esta limpia en magnitud.

```python
df = df.dropna(subset=['Total Deaths'])
df.info()
```

Una vez limpia, procedemos a filtrar los años que vamos a utilizar en este caso del 2016 a 2018

```python
df = df[df["Start Year"] >= 2016] #Año inicial
df = df[df["Start Year"] <= 2018] #Año final

df.info()
```

Como se pudo observar pasamos a tener solamente 97 filas de datos, asi que acomodamos el indice para mantener el orden en nuestra nueva base de datos.

```python
indices = list(range(1,98))
df.index = indices
df.head()
```

Guardamos la data procesada de magnitud-muertos 2016-2018

```python
df.to_excel("/content/drive/MyDrive/Colab Notebooks/Magnitud_Muertos_2016_2017_2018.xlsx")
```

# Graficas y Analisis

Para graficar importamos las librerias que utilizaremos

```python
import matplotlib.pyplot as plt
```

En las graficas que se mostraran a continuacion se dara a conocer de manera general los datos que se recolectaron sobre los desastres hidrologicos.



1. Grafica de desastres hidrologicos por region


```python
# Gráfica de desastres hidrologicos por región
region = df["Region"]
plt.hist(region, color="purple")
plt.title("Desastres por región")
plt.ylabel("Cant. de desastres")
plt.xlabel("Región")
plt.show()
```

Como se pudo obervar la region que obtuvo mas desastres hidrologicos fue Asia con mas de 40 cantidad de desastres.


2.   Grafica de desastres por año



```python
año = df["Start Year"]
plt.hist(año, color="Purple")
plt.title("Desastres por año")
plt.ylabel("Cant. de desastres")
plt.xlabel("Año")
plt.xticks(año.unique(), rotation=45)
plt.show()
```

a lo largo de estos tres años el 2016 fue el que tuvo mas cantidad de desastres hidrologicos, pero por lo que podemos obsvervar fue disminuyendo en cantidad hasta llegar al 2018.




3.   Grafica de muertes totales por pais



```python
# Gráfica de muertes totales por país.
plt.figure(figsize=(10, 5))
plt.bar(df['Country'], df['Total Deaths'], color = "purple")
plt.title('Muertes Totales por País')
plt.xlabel('Paises')
plt.ylabel('Total de muertos')
plt.xticks(rotation=90)
plt.show()
```

En esta grafica determinamos que los 6 paises con mas fatalidad fueron:


1.   China
2.   Japan
3.   Peru
4.   Nepal
5.   Corea del Norte
6.   Nigeria



donde cuatro de ellos son de la region de asia, siendo esta la region con mas fatalidad.




4.   Grafica de total de muertos




```python
# Gráfica de total de muertos

total_deaths = df['Total Deaths']

plt.figure(figsize=(14,8))
plt.plot(total_deaths, marker='D',linewidth=2, markersize=5)
plt.title('Total de muertos', fontsize=16)
plt.xlabel('Cant. de datos', fontsize=14)
plt.ylabel('Cant. de Muertos', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```

comparando la grafica anterior con esta entendemos que la mayor cantidad de muertos fue llegando casi 300 muertos en una sola inundacion, la cual sabemos perfectamente que tuvo protagonismo en china.

```python
total_deaths.describe()
```



5.  Grafica de Magnitud del desastre



```python
magnitud = df['Magnitude']
plt.figure(figsize=(14,8))
plt.plot(magnitud, marker='D',linewidth=2, markersize=5)
plt.title('Magnitud del desastre', fontsize=16)
plt.xlabel('Cant. de datos', fontsize=14)
plt.ylabel('Magnitud', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

Para finalizar medimos la mayor magnitud que llegaron a tener dichas inundaciones, la mas alta fue casi llegando a 2.5.

```python
magnitud.describe()
```



6.   Grafica de Relacion entre la magnitud-muertes




```python
relacion_mag_deaths = pd.DataFrame({'Magnitud':df['Magnitude'],  'Muertos':df['Total Deaths']})
relacion_mag_deaths = relacion_mag_deaths.sort_values(by='Muertos')
relacion_mag_deaths.head()
plt.figure(figsize=(14,8))
plt.scatter(relacion_mag_deaths['Magnitud'], relacion_mag_deaths['Muertos'])
plt.show()
```



7.   Mapa de calor (Correlacion)



```python
import seaborn as sns
# Matriz de correlación
correlación = df[['Total Deaths', 'Magnitude']].corr()
sns.heatmap(correlación, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
